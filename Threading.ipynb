```python
import time
def greet1():
    for i in range(1000):
        print("hello ")
        time.sleep(1)
def greet2():
    for i in range(100):
        print("hi")
        time.sleep(2)
```

```python
t1=Thread(target=greet1,name="hello thread")
t2=Thread(target=greet2,name="hi thread")
```

```python
t1.start()
t2.start()
```

```python
greet1()
greet2()
```
